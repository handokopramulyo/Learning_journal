```python
# define dataframe
data = pd.DataFrame({'sisi1':[5,8,2,5,7,15,6,20,56,1,42,7],'sisi2':[15,82,12,5,6,15,6,4,27,2,4,9]})
data
```

```python
# Buat fungsi tuk hitung luas persegi dan jajargenjang

# hitung luas persegi
def hasillp(p,l):
    rumuslp = p*l
    return rumuslp

# hitung luas jajargenjang
def hasiljg(t,u):
    rumusjg = t*u*1000 # untuk pembeda supaya terlihat
    return rumusjg
```

```python
length = range(len(data['sisi1']))
length
```

```python
# penggunaan dataframe di dalame for, if, loc/iloc
# operasikan value kisi1*kisi2 >= 10 ke fungsi luas persegi,
# else, operasikan kisi1*kisi2 < 10 ke fungsi luas jajargenjang

#### Gaveup !!!! Need Help

for row in length:
    if data['sisi1'][row]>=10:
        kolom3 = hasillp(data['sisi1'][row],data['sisi2'][row])
    else:
        kolom3 = hasiljg(data['sisi1'][row],data['sisi2'][row])
```
